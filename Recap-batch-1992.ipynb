### ‚ùì 1 - Draw your first map

The goal of this first question is to:

- Draw a map centered at specific coordinates (check out [gps-coorinates.net](https://www.gps-coordinates.net/))
- Put a marker at that specific point
- Draw a circle of 100km centered at specific point

Check out the [`folium` Quickstart](https://python-visualization.github.io/folium/quickstart.html), you should find all the code examples you need, good luck!

````python
coordinates = [48.8649224, 2.3800903] # 16 Villa Gaudelet, Paris

m = folium.Map(
    location=coordinates,
    zoom_start = 8
)

folium.Marker(
    location=coordinates,
    tooltip="Click me!",
    popup="Timberline Lodge",
    icon=folium.Icon(color="green"),
).add_to(m)

radius = 100_000
folium.Circle(
    location=coordinates,
    radius=radius,
    color="black",
    weight=1,
    fill_opacity=0.2,
    opacity=1,
    fill_color="blue",
    fill=False,  # gets overridden by fill_color
    popup="{} meters".format(radius),
    tooltip="I am in meters",
).add_to(m)

m
````

### ‚ùì 2 - Build a User-Friendly tool

The problem of the previous code is that we need to manually convert an address into the GPS coordinates. Let's implement a program so that user can specify their plain text address and get the map drawn for that point:

```python
draw_100km_map("138 Kingsland Rd, London E2 8DY")
```

üí° Hint: You can use the [Nominatim Search API](https://nominatim.org/release-docs/latest/api/Search/) to implement a `geocode(address)` function which returns a list of `[lat, lng]` coordinates. 

üí° Getting a 403 Forbidden error? The Nominatim API is free to use, but doesn't accept the default headers that most http packages use. So you have to provide your own headers when making the request. Giving a name for your app in `User-Agent` should be enough.

````python
!curl "https://nominatim.openstreetmap.org/search?q=20 rue des capucins, Lyon&format=json"
````

````python
import requests

url = "https://nominatim.openstreetmap.org/search"

response = requests.get(url, params = {"q": "rue des capucins, Lyon", "format": "json"}, headers = {'User-Agent': 'test_app'})
response.json()[0]
````

````python
def draw_100km_map(user_input):
    # user inout > coordonn√©es 
    response = requests.get(url, params = {"q": user_input, "format": "json"}, headers = {'User-Agent': 'test_app'})
    response = response.json()[0]

    coordinates = [response['lat'],response['lon']]

    
    # injecter coord dans le code poyur g√©n√©rer map
    m = folium.Map(
        location=coordinates,
        zoom_start = 8
    )
    
    folium.Marker(
        location=coordinates,
        tooltip="Click me!",
        popup="Timberline Lodge",
        icon=folium.Icon(color="green"),
    ).add_to(m)
    
    radius = 100_000
    folium.Circle(
        location=coordinates,
        radius=radius,
        color="black",
        weight=1,
        fill_opacity=0.2,
        opacity=1,
        fill_color="blue",
        fill=False,  # gets overridden by fill_color
        popup="{} meters".format(radius),
        tooltip="I am in meters",
    ).add_to(m)

    return m
````

The interactive tool:

````python
address = input("Input your address > ")
draw_100km_map(address)
````
